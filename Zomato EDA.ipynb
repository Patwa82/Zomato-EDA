```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
## to check data types 
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## pie chart  top 3 country that use for zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

observation:Zomato maximum record transacton are from india after that USA and then United Kingdm

```python
## Numerical variables
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
```

```python
ratings
```

## observation
1. when rating is between 4.5 to 4.9 -----> Excellent
2. when rating is between 4.0 to 3.4 -----> very good
3.  when rating is between 3.5 to 3.9 -----> Good
4.  when rating is between 3.0 to 3.4 ------> average
5.  when rating is between 2.5 to 2.9 -----> average
6.  when rating is between 4.0 to 2.4 ----> poor

```python
ratings
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Counts',data=ratings)
plt.show()
```

```python
sns.barplot(x='Aggregate rating',y='Rating Counts',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation 
1. Not rated count is very high 
2. Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green'])
```
